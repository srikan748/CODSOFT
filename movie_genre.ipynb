```python
import pandas as pd
train_df = pd.read_csv('train_data.txt',sep=':::',engine='python',names=['ID', 'Title', 'Genre', 'Description'])
print(train_df.head())
```

```python
train_df['Text'] = train_df['Title'] + ""+train_df['Description']
x=train_df['Text']
y=train_df['Genre']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)
x_vect = vectorizer.fit_transform(x)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_vect, y)
```

```python
test_df = pd.read_csv('test_data.txt', sep=':::', engine='python', names=['ID', 'Title', 'Description'])
test_df['Text'] = test_df['Title'] + " " + test_df['Description']
X_test_vect = vectorizer.transform(test_df['Text'])
```

```python
predictions = model.predict(X_test_vect)
test_df['Predicted Genre'] = predictions
print(test_df[['ID', 'Predicted Genre']])
```
