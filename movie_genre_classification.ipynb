1.READ DATA

```python
import pandas as pd                   #import pandas
import warnings                       #import warnings and to ignore warnings
warnings.filterwarnings("ignore")     
data=pd.read_csv("description.txt")  #read data
data                                 #display data
```

2.CREATE AND APPLY FUNCTION TO READ DATA BY SPLITTING

```python
def load_data(file_path):                                     #function that takes file path
    with open(file_path,'r',encoding='utf-8') as f:           #read data in read mode with utf-8 encoding refernce f
        data=f.readlines()                                    #take each line
        data = [line.strip().split(' ::: ') for line in data]  #split each line
        return data                                             #return splitted line
```

```python
train_data = load_data("train_data.txt")        #to load training dataset and pass to load data function
train_df = pd.DataFrame(train_data,columns=['ID','Title','Genre','Description']) #convert to dataframe and rename columns

test_data = load_data("test_data.txt")                                         #to load dataset and pass to load data function
test_df = pd.DataFrame(test_data,columns=['ID','Title','Description'])      #convert to dataframe and rename columns

test_solution = load_data("test_data_solution.txt")                      # #to load dataset and pass to load data function
test_solution_df = pd.DataFrame(test_solution,columns=['ID','Title','Genre','Description'])  #convert to dataframe and rename columns


```

```python
print("Train Data:")
train_df
```

```python
print("\nTest Data:")
test_df
```

```python
print("\nTest solution:")
test_solution_df
```

3.FEATURE EXTRACTION : TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer    #import  TfidfVectorizer  
vectorizer = TfidfVectorizer(max_features=10000)              #initailise tfidf with 10k features

x_train_tfidf = vectorizer.fit_transform(train_df["Description"]) #fit and transform training dataframe column named description
x_test_tfidf = vectorizer.transform(test_df["Description"])       #fit and transform training dataframe column named description

print(f"training data shape: {x_train_tfidf.shape}")            #display shape of training dataset
print(f"test data shape: {x_test_tfidf.shape}")                 #display shape of training dataset
```

4.ENCODING TARGET LABELS

```python
from sklearn.preprocessing import LabelEncoder     #importing  LabelEncoder  
label_encoder = LabelEncoder()                    #encoding values into numerical labels
y_train = label_encoder.fit_transform(train_df['Genre'])  #encoding dataframe column into numerical labels
print(f"unique genres in the training data: {label_encoder.classes_}") #check unique genres after encoding
```

5.MODEL BUILDING-LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression

# Train logistic regression model
lr_model = LogisticRegression(max_iter=200)
lr_model.fit(x_train_tfidf, y_train)

# Predict on test data
y_pred = lr_model.predict(x_test_tfidf)

# Convert numeric labels back to original genres
predicted_genres = label_encoder.inverse_transform(y_pred)

# Add predictions to DataFrame
test_df['predicted_Genre'] = predicted_genres

# Display Title and predicted Genre
test_df[['Title', 'predicted_Genre']]
```

```python
test_df['predicted_Genre'] = predicted_genres  #adding predicted genres to test dataframe

merged_df = pd.merge(test_solution_df[['ID','Genre']],test_df[['ID','predicted_Genre']],on='ID')  #now merging
merged_df      #check columns of merged_df to confirm predicted_genre present
```

6.MODEL EVALUATION-LOGISTIC REGRESSION

```python
from sklearn.metrics import accuracy_score,classification_report     #import modulus for metric calculation
accuracy = accuracy_score(merged_df['Genre'],merged_df['predicted_Genre'])  #get accuracy
print(f"Accuracy:{accuracy:.4f}")                                            #print accuracy with 4 decimal points
print("\nclassification Report:")                                            #to get classification report
print(classification_report(merged_df['Genre'],merged_df['predicted_Genre']))  #display and print classification report
```

7.MODEL BUILDING - NAIVE BAYES

```python
from sklearn.naive_bayes import MultinomialNB  #import MultinomialNB 
nb_model = MultinomialNB()                      #building naive bayes model
nb_model.fit(x_train_tfidf,y_train)             #training naive bayes model
```

```python
y_pred_nb = nb_model.predict(x_test_tfidf)                      #predicting the genres on test dataset
predicted_genres_nb = label_encoder.inverse_transform(y_pred_nb)   #applying inverse transform on predicted values
test_df['predicted_Genre_NB'] = predicted_genres_nb                 #add predictions to the test dataframe
merged_df_nb = pd.merge(test_solution_df, test_df[['ID', 'predicted_Genre_NB']], on='ID')  #merge dataframe
```

8.MODEL EVALUATION - NAIVE BAYES

```python
from sklearn.metrics import accuracy_score,classification_report   # import accuracy_score,classification_report
accuracy_nb = accuracy_score(merged_df_nb['Genre'],merged_df_nb['predicted_Genre_NB']) #finding accuracy
print(f"Naive Bayes Accuracy:{accuracy_nb}")
print("Naive Bayes classification Report:")            #finding classification report
print(classification_report(merged_df_nb['Genre'],merged_df_nb['predicted_Genre_NB'],target_names=label_encoder.classes_))
```
